```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
steps = [i for i in np.arange(0, 1e4, 200)]
```

```python
def learning_rate_decay(step, lr=0.002, decay=0.0002):
    
    return lr / (1 + decay * step)

def learning_rate_decay_2(step, lr=0.003, decay=0.0005):
    
    return lr / (1 + decay * step)

def learning_rate_decay_3(step, lr=0.003, decay=0.0002):
    
    return lr / (1 + decay * step)
```

```python
value = list(map(learning_rate_decay, steps))
value_2 = list(map(learning_rate_decay_2, steps))
value_3 = list(map(learning_rate_decay_3, steps))

plt.plot(value)
plt.plot(value_2)
plt.plot(value_3)

plt.show()
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
x = np.random.random(100)/10 + 1
xx = np.random.random(100)/10 + 2
y = np.random.random(100)

z = np.asarray([x, xx, y])
z = z.T
z_1 = z

x = np.random.random(100)/10 - 1
xx = np.random.random(100)/10 + 2
y = np.random.random(100)

z = np.asarray([x, xx, y])
z = z.T
z_2 = z

z = np.concatenate([z_1, z_2])
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(z)
```

```python
z_t = (pca.transform(z))
plt.plot(z_t[:, 0], z_t[:, 1], '.')
```

```python
z_t = TSNE(n_components=2).fit_transform(z)
plt.plot(z_t[:, 0], z_t[:, 1], '.')
```
